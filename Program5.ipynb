# while loop

```python
a = 3
while (a>=1):
  print(a)
  print('Hello')
  a -= 1
```

### Type of while loop
1. Infinite loop
2. while with else statement
3. single statement

# Infinite loop

```python
a = 5
while (a>=5):
  print(a)
  a += 1
```

# while with else
- when the condition become false, the block under else statement will be executated.

```python
a = 3
while (a>0):
  print(a)
  a -= 1
else:
  print("reach at 0")
```

# single statement while loop
- in single statement while loop ';' is optional

```python
a = 3
while (a>0):print(a);a-=1;
```

```python
a = 3
while (a>0):print(a);a-=1
```

```python
# In single statement while with else is possible: Yes, it's possible
i = 0
while i < 5: i += 1
else: print("Loop completed normally")
```

# For loop
- it can iterate for sequency of pattern
- we use 'in' key word
- 'in' is membership operator
- range function use for iteration
- we can also iterate in for loop without using range function

```python
range (10)
```

```python
list(range(10))
# we use list function to convert range object into list object
```

```python
list(range(2,10))
```

```python
list(range(2,10,2))
```

```python
list(range(10,2))
```

```python
list(range(2,10,-2))
```

```python
list(range(12,2,-2))
```

```python
for i in range(7):
  print(i)
```

```python
# how to iterate in for loop with out using range function
for i in [1,5,9,55,100]:
  print(i)
```

```python
for i in "Python":
  print(i)
```

```python
for i in {2,5,3,3,7}:
  print(i)
```

```python
# iterate over indices
l1 = ["c","c++","java","python"]
for i in range(len(l1)):
  print(i)
```

```python
l1 = ["c","c++","java","python"]
for i in range(len(l1)):
  print(l1[i])
```

```python
for i in l1:
  print(i)
```

```python
for i in range(7):
  print(i)
else:
  print("reach at else")
```

```python
for i in range(7):
  print(i)
else:
  print("reach at else")
  print(i)
```

# Nested for loop

```python
for i in range(5):
  for j in range(5):
    print(i,j)
```

```python
for i in range(5):
  for j in range(i):
    print('*')
```

```python
for i in range(5):
  for j in range(i):
    print('*',end='')
```

```python
for i in range(5):
  for j in range(i):
    print('*',end='')
  print()
```

```python
for i in range(5):
  for j in range(i):
    print('*',end=' ')
  print()
```

```python
for i in range(6):
  for j in range(i):
    print('*',end=' ')
  print()
```

```python
i=0
j=0
while (i<6):
  j=0
  while (j<i):
    print('*',end=' ')
    j+=1
  i+=1
  print()
```

```python
i=0
while (i<6):
  j=0
  while (j<i):
    print('*',end=' ')
    j+=1
  i+=1
  print()
```
